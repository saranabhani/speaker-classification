```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Path to your text file
file_path = 'confusion_matrix_with_accuracy.csv'

confusion_matrix = pd.read_csv(file_path, sep=',', index_col=0).drop(columns=['Accuracy'])
confusion_matrix = confusion_matrix.values


# Normalizing the confusion matrix
confusion_matrix_normalized = confusion_matrix.astype('float') / confusion_matrix.sum(axis=1)[:, np.newaxis]
```

```python
plt.figure(figsize=(10, 10)) 
# Create a heatmap of the confusion matrix
plt.imshow(confusion_matrix_normalized, interpolation='none', cmap='magma_r')
plt.colorbar()

plt.xticks([])
plt.yticks([])
plt.title('Confusion Matrix Heatmap')

# Show the heatmap
plt.show()
```

```python
confusion_matrix = pd.read_csv(file_path, sep=',', index_col=0)

# Plotting a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(confusion_matrix['Accuracy'], vert=False, patch_artist=True)

# Adding labels and title
plt.xlabel('Accuracy')
plt.title('Distribution of Accuracy Across Speakers')
plt.yticks([1], ['Speakers'])

# Show the plot
plt.show()
```
